```python
sal = np.array([12000,60000,80000,76000])
wd = np.array([210,300,230,256])
years = [1,2]
bonus=[]
for i,j in zip(sal,wd):
  if j > 250:
    if i > 0:
      i += i * 0.1
      bonus.append(i)
    else:
      i -= i * 0.1
      bonus.append(i)
print(bonus)
```

```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '7',label =
         'CompanyA',marker='.',ms='20',mec='r')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label=
'CompanyB',marker='H')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# load example data set
tips = sns.load_dataset("tips")
print(tips)
# creat a scatter plot
sns.scatterplot(x="total_bill",y="tip",data=tips,label='bill')
plt.legend()
plt.title("scatter plot of total bill vs,tip")
plt.xlabel("total Bill ($)")
plt.ylabel("tip ($)")
plt.show()
```
